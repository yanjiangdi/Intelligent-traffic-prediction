```python
import numpy as np
import matplotlib.pyplot as plt
from operator import itemgetter, attrgetter
import time,datetime
from sklearn.cross_validation import train_test_split
import pandas as pd
from pandas import DataFrame

import h2o
h2o.init()
from h2o.estimators.random_forest import H2ORandomForestEstimator 


def get_week_day(d):
    date=d.split('-')
    year=int (date[0])
    month=int(date[1])
    day=int(date[2])
    
    days=[31,28,31,30,31,30,31,31,30,31,30,31]
    if (year%4==0 and  year%100!=0) or (year%400==0):
        days[1]=29
    else:
        days[1]=28
    sum_day=0 
    for i in range(0,month-1):
        sum_day+=days[i]
    sum_day+=day
    return (year-1+int((year-1)/4)-int((year-1)/100)+int((year-1)/400)+sum_day)%7


df1=pd.read_table('gy_contest_link_info.txt', header=0,sep=';', encoding='utf-8')
df2=pd.read_table('gy_contest_link_top.txt', header=0,sep=';', encoding='utf-8')
df_t=pd.read_table('test.txt', header=0,sep=';', encoding='utf-8')

r0=pd.read_table('0.csv', header=0,sep=',', encoding='utf-8')
r1=pd.read_table('1.csv', header=0,sep=',', encoding='utf-8')
r2=pd.read_table('2.csv', header=0,sep=',', encoding='utf-8')
r3=pd.read_table('3.csv', header=0,sep=',', encoding='utf-8')
r4=pd.read_table('4.csv', header=0,sep=',', encoding='utf-8')
r5=pd.read_table('5.csv', header=0,sep=',', encoding='utf-8')
r6=pd.read_table('6.csv', header=0,sep=',', encoding='utf-8')


df2=df2.fillna(0)
df2['inN']=0
df2['outN']=0
for i in range(len(df2)):
    inlink=df2.iat[i,1]
    outlink=df2.iat[i,2]
    if inlink==0:
        df2.iat[i,3]=0
    else:
        df2.iat[i,3]=len(inlink.split("#"))
    if outlink==0:
        df2.iat[i,4]=0
    else:
        df2.iat[i,4]=len(outlink.split("#"))
df4=pd.merge(df1,df2)
df_test=pd.merge(df4,df_t)

df_test['week']=df_test['date'].apply(lambda x:get_week_day(x))

df_test_0=df_test[df_test['week']==0]
df_test_1=df_test[df_test['week']==1]
df_test_2=df_test[df_test['week']==2]
df_test_3=df_test[df_test['week']==3]
df_test_4=df_test[df_test['week']==4]
df_test_5=df_test[df_test['week']==5]
df_test_6=df_test[df_test['week']==6]

df_test_0['travel_time']=0.0
df_test_0=pd.DataFrame(df_test_0)
for i in range(0,len(df_test_0)):
    df_test_0.iat[i,11]=float(r0.iat[i,0])
    
df_test_1['travel_time']=0.0
df_test_1=pd.DataFrame(df_test_1)
for i in range(0,len(df_test_1)):
    df_test_1.iat[i,11]=float(r1.iat[i,0])

df_test_2['travel_time']=0.0
df_test_2=pd.DataFrame(df_test_2)
for i in range(0,len(df_test_2)):
    df_test_2.iat[i,11]=float(r2.iat[i,0])
    
df_test_3['travel_time']=0.0
df_test_3=pd.DataFrame(df_test_3)
for i in range(0,len(df_test_3)):
    df_test_3.iat[i,11]=float(r3.iat[i,0])
    
df_test_4['travel_time']=0.0
df_test_4=pd.DataFrame(df_test_4)
for i in range(0,len(df_test_4)):
    df_test_4.iat[i,11]=float(r4.iat[i,0])
    
df_test_5['travel_time']=0.0
df_test_5=pd.DataFrame(df_test_5)
for i in range(0,len(df_test_5)):
    df_test_5.iat[i,11]=float(r5.iat[i,0])

df_test_6['travel_time']=0.0
df_test_6=pd.DataFrame(df_test_6)
for i in range(0,len(df_test_6)):
    df_test_6.iat[i,11]=float(r6.iat[i,0])
    

df_test_0[['link_ID','date','time_interval','travel_time']].to_csv('r0.txt',sep=';')

df_test_1[['link_ID','date','time_interval','travel_time']].to_csv('r1.txt',sep=';')

df_test_2[['link_ID','date','time_interval','travel_time']].to_csv('r2.txt',sep=';')

df_test_3[['link_ID','date','time_interval','travel_time']].to_csv('r3.txt',sep=';')

df_test_4[['link_ID','date','time_interval','travel_time']].to_csv('r4.txt',sep=';')

df_test_5[['link_ID','date','time_interval','travel_time']].to_csv('r5.txt',sep=';')

df_test_6[['link_ID','date','time_interval','travel_time']].to_csv('r6.txt',sep=';')







```
